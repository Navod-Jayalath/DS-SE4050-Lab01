```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv('Au_nanoparticle_dataset.csv')
df.head()
```



Question 1

```python
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
new_df.head()
```

Question 2

```python
new_df.head(20)
```

Question 3

```python
# Mean
print("Mean:\n", new_df.mean())

# Standard Deviation
print("\nStandard Deviation:\n", new_df.std())

# Quartiles (25%, 50%, 75%)
print("\nQuartiles:\n", new_df.quantile([0.25, 0.5, 0.75]))
```

Question 4

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 4))

for i, col in enumerate(new_df.columns):
    plt.subplot(1, 4, i + 1)
    plt.hist(new_df[col], bins=30, color='skyblue', edgecolor='black')
    plt.title(col)
    plt.xlabel('Value')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Question 5

```python
import seaborn as sns

sns.pairplot(new_df)
```

Question 6

```python
g = sns.PairGrid(new_df)

# Upper: bivariate histograms
g.map_upper(sns.histplot)

# Diagonal: histogram + KDE
g.map_diag(sns.histplot, kde=True)

# Lower: bivariate KDE plots
g.map_lower(sns.kdeplot, fill=True)

plt.show()
```
